```python
import pandas_datareader as wb

data_frame = pd.DataFrame()

tickers_array = []

for t in tickers:
    tickers_array.append(t)
    
i = 1

for t in tickers_array:
    data_frame[t] = wb.DataReader(t, data_source = "iex", start = "2013-04-18")["close"]
    print(str(i), t)
    i += 1
    
data_frame
```

```python
returns = (data_frame / data_frame.shift(1)) - 1
returns
```

```python
(data_frame / data_frame.iloc[0] * 100).plot(figsize=(15,8))
plt.show()
```

```python
#put all of this in a for loop that iterates the day
#closing in on the date from left and right and then checking the middle out
starting_date = "2018-07-20"

avg_annual_returns = returns.loc[starting_date:].mean() * 250

print(len(avg_annual_returns))

double_returns = avg_annual_returns.to_dict()

for key, n in avg_annual_returns.to_dict().items():
    if(n < 2 * avg_annual_returns['SPY']):
        del double_returns[key]
        
print(double_returns)  
print('')
print(len(double_returns))
    
```

```python
from datetime import date, timedelta

#fix to to start from today and count down!
d1 = date(2013, 4, 18)  # start date
d2 = date(2018, 7, 19)  # end date

delta = d2 - d1         # timedelta

time_data = pd.Series()

for i in range(delta.days + 1):
    avg_annual_returns = returns.loc[str(d1 + timedelta(i)):].mean() * 250

    double_returns = avg_annual_returns.to_dict()

    for key, n in avg_annual_returns.to_dict().items():
        if(n < 2 * avg_annual_returns['SPY']):
            del double_returns[key]
    
    time_data[(delta.days - i)/30] = len(double_returns) / len(avg_annual_returns) * 100
  
#print(time_data)
time_data.plot(figsize=(15,8))
plt.ylabel('% of stocks in MTUM outpreforming S&P by 2x or more')
plt.xlabel('Months from 7/19/18')
plt.axis([1, 60, 0, 100])
plt.show()
```
