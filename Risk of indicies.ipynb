```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
from alpha_vantage.timeseries import TimeSeries
```

```python


tickers = ["SPY", "MTUM"]
ts = TimeSeries(key = 'BEII34ZY6UI335CP', output_format = 'pandas')
data_source = pd.DataFrame()
for t in tickers:
    INDI_av, metadata = ts.get_daily_adjusted(t, outputsize="full")
    data_source[t] = INDI_av.loc['2013-04-18':]["5. adjusted close"]
data_source
```

```python
(data_source / data_source.iloc[0] * 100).plot(figsize=(15,6))
```

```python
returns = (data_source / data_source.shift(1)) - 1
returns
```

```python
daily_returns = returns.mean()
daily_returns
```

```python
annual_returns = returns.mean() * 250
annual_returns
```

```python
vols = returns[["SPY", "MTUM"]].std() * 250 ** 0.5
vols["MTUM"] / vols["SPY"]
```

```python
var = returns[["SPY", "MTUM"]].var() * 250
var
```
